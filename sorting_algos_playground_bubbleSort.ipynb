```python
# Bubble Sort
# Here I'm making the funciton to use in the viuslisation
# THE ACTUALLY GOOD ONE, W/OUT PRINT DEBUG STATEMENTS

# This version also displays the bar values and shows whats being compared
def bubble_sort_visual(lst):
  # For the function logic to compleatly sort the list, it needs to repeat the
  # sort multiple times.

  # The maximum number of swaps we need to perform is the lenght of the list,
  # i.e. where we've checked every element in the list. So, we need to run
  # the function for as many items in the list.

  itterations = len(lst)

  # For the visualisation we'll create a state log list that will sorte the
  # output for every itteration. We can when loop through this list for
  # the visualisation
  states_log = []

  # Creating a list to keep ttrack of whats being compared in each iteration
  comparison_log = []

  for iteration in range(itterations):
    # Create a starting pointer index to use to access the values in the list
    pos = 0

    # To ensure our pointer dose not exceed the range of list elments
    while pos < len(lst) - 1:
      # Get the first value to compare
      idx_one = pos
      val = lst[idx_one]

      # Get the next value in the list
      idx_two = pos + 1
      val_two = lst[idx_two]

      # Compare the two values
      if val > val_two:
        lst[idx_two] = val
        lst[idx_one] = val_two
        # Track the fact we've performed a swap

      # Go to next value in list
      pos += 1

      # Capture state of itteration
      states_log.append(list(lst))
      comparison_log.append((idx_one, idx_two))

  return states_log, comparison_log


# Generate random list
lst = [random.randrange(1,10) for x in range(1, 20)]
print(f'List to sort: {lst}')

# Call bubble sort function
states_log, comparisons_log = bubble_sort_visual(lst)

app2 = Dash(__name__)

app2.layout = html.Div([
    dcc.Graph(id='live-graph', animate=True),
    dcc.Interval(
        id='interval-component',
        interval= 700,  # in milliseconds
        n_intervals=0
    )
])

# Callback to update graph
@app2.callback(
    Output('live-graph', 'figure'),
    [Input('interval-component', 'n_intervals')]
)
def update_graph2(n):
    step = n % len(states_log)
    current_state = states_log[step]
    idx_one, idx_two = comparisons_log[step]

    fig = px.bar(x=list(range(len(current_state))),
                 y=current_state,
                 labels={'x':'Index', 'y':'Value'},
                 text=current_state)

    fig.update_traces(marker_color=['red' if i == idx_one or i == idx_two else 'blue' for i in range(len(current_state))])

    fig.update_layout(
        title="Bubble Sort Step: {}".format(step),
        showlegend=False,
        yaxis=dict(range=[0, max(current_state) + 2])
    )

    return fig

if __name__ == '__main__':
    app2.run_server()
```
