```python
# Quick Sort

# List to be sorted
LST = [8,3,6,1,4,6,3,5,67,75,64,543,6]


# Quick sort function

def quick_sort(lst):
  # If list contains 1 or less elements then the list is already sorted
  if len(lst) <= 1:
    return lst

  # When list is more than 1 element
  else:
    print(f'List to sort {LST}')

    # Define pivot point
    # Defined as last element in list
    pivot_point = len(lst) - 1
    print(f'Pivot point defined as index position {pivot_point}')

    # Value of pivot element in list
    pivot = lst[pivot_point]
    print(f'Value of pivot point {pivot}')

    # The pivot element allows us to split the orignal list
    # into two sub arrays

    # Defining low and high variables
    # low is defined as the first element in the list
    # high is defined as second-to-last index value
    # we use the second to last since the last value is our
    # pivot point
    low = 0
    high = len(lst) - 2

    # This condition ensures that the loop continues until the low
    # index crosses the high index.
    while low <= high:

      # if the first element in the list is less than or equal to the
      # pivot point element in the list, then incease low by 1
      if lst[low] <= pivot:
        low += 1

      # if the last element in the list is lessthan or equal the pivot point
      # element then decrease high by 1
      elif lst[high] > pivot:
        high -= 1

      else:
        # Swap the low and high index values from list
        # Values are swaped to ensure that elements less than or equal to
        # the pivot point are placed on the left hand side
        # and elements greater than the pivot are placed on the right hand side
        lst[low], lst[high] = lst[high], lst[low]
        low += 1
        high -= 1

    # Swap pivot point with the element at the low index point
    lst[low], lst[-1] = lst[-1], lst[low]

    # Recursivly sort the left and right sub arrays
    print(f'left array: {lst[:low]}')
    print(f'right array: {lst[low+1:]}')
    left = quick_sort(lst[:low])
    right = quick_sort(lst[low+1:])

    # Combine sorted sub arrays and pivot point into a single list
    return left + [lst[low]] + right

sorted_lst = quick_sort(LST)
print(sorted_lst)
```
